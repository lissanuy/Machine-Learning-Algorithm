#**Model Regresi Linier Sederhana**

Regresi linier adalah metode untuk mendefinisikan hubungan linier antara variabel dependen **Y** dan variabel independen **X**. Secara sederhana dapat dituliskan dalam formula berikut:

 $Y_i = b_0 +b_1X_i + ϵ_i $

Yang mana $Y_i$ adalah variabel dependen, $X_i$ adalah variabel independen ,ϵ adalah kesalahan acak, $b_0$ disebut sebagai intercept dan $b_1$ disebut sebagai koefisien regresi yang dapat diestimasi berdasarkan data $Y_i$ dan $X_i$ yang diberikan, Koefisien $b_0$ juga biasa disebut sebagai bias, tujuan utama dari model ini adalah untuk mengambarkan garis linier dengan kesesuain terbaik (best fit) antara X dan Y yang mengestimasi hubungan antara X dan Y.


#**Metode Gradient Descent**
 Sebelumnya kita telah membahas tentang estimasi hubungan X dan Y pada sebuah garis. Misalnya, kita sediakan input dan output sampel, kemudian kita memplot titik-titik data ini pada grafik 2 dimensi.

Dan kesalahan total model linier adalah jumlah kesalahan dari setiap titiknya, yaitu:


$\sum \limits _{i=1}^{n}{{\epsilon}_{i}}^2 $

$\epsilon_{i}$ = Jarak antara garis ke titik data ke-i
n= jumlah total titik data

Formula di atas diperoleh dari:

$y_i=b_0+b_1.x_i+_i$

$\epsilon_{i}= y_i-b_0-b_1.x_i$

$\sum \limits_{i=1}^{n} {{\epsilon}_{i}}^2 = \sum\limits_{i=1}^{n} (y_i-b_0-b_1.x_i)^2$

JIka pada pembahasan sebelumnya kita mengestimasi koefisien $b_0$ dan $b_1$ menggunakan metode OLS. Maka kita akan menggunakan metode lain yaitu Gradient Descent. Ada catatan penting ketika kita menerapkan metode OLS saat ingin mengestimasi nilai parameter koefisien $b_0$ dan $b_1$, yaitu ketika nilai turunan fungsi error diasumsikan sama dengan nol saat mencapai nilai minimum.

Faktanya, nilai minimum Global yang dicari tidak selalu terjadi ketika turunan pertamanya sama dengan nol. Inilah yang menjadi ide dasar dari metode selanjutnya yaitu Gradient Descent. Alih-alih langsung mengasumsikan turunan pertama dari fungsi error terhadap parameter adalah sama dengan nol. Maka metode ini justru mengggunakan formula turunan parsial pertama dari fungsi error untuk memperbaharui nilai estimasi parameter koefisien $b_0$ dan $b_1$. Proses ini diulangi(iterasi) hingga memperoleh kriteria yang diharapkan.

Pembaharuan dari nilai parameter $b_0$ sdan $b+1$ adalah sebagai berikut:

$b_0(baru)=b_0(lama)- \Delta{b_0} \times tk$

$b_1(baru)=b_1(lama)- \Delta{b_1} \times tk$

yang mana $tk$ adalah tingkat ketelitian pembelajaran model ML atau ```learning rate.```

Pada notebook ini kita juga akan menggunakan fungsi error(Cost) yang berbeda yaitu MSE(Mean Square Error) dengan formula seperti di bawah ini:

$Cost=\frac{1}{n} \sum\limits_{i=1}^{n} (y_i-(b_0+b_1.x_i))^2 $

$\Delta{b_0}$ dan $\Delta{b_1}$ adalah nilai perubahan parameter $b_0$ dan $b_1$ yang diperoleh dari turunan parsial pertama fungsi Cost yaitu sebagi berikut:

$\Delta{b_0}=\frac{\partial Cost}{\partial b_0}$

$=\frac{1}{n} \sum\limits_{i=1}^{n} 2(y_i-(b_0+b_1\times x_i))(-1) $

$=\frac{-2}{n} \sum\limits_{i=1}^{n} (y_i-(b_0+b_1\times x_i))$

$=\frac{-2}{n} \sum\limits_{i=1}^{n} (y-y') $



$\Delta {b_1}=\frac{\partial Cost}{\partial b_1}$

$=\frac{1}{n} \sum\limits_{i=1}^{n} 2(y_i-(b_0+b_1\times x_i))(-1) x $

$=\frac{-2}{n} \sum\limits_{i=1}^{n} (y_i-x.(b_0+b_1\times x_i))$

$=\frac{-2}{n} \sum\limits_{i=1}^{n} x.(y-y') $


Sehingga pembaharuan nilai parameter $b_0$ dan $b_1$ adalah sebagai berikut:

$b_0(baru)=b_0(lama)- \frac{-2}{n} \sum\limits_{i=1}^{n} (y-y') \times tk$

$b_1(baru)=b_1(lama)- \frac{-2}{n} \sum\limits_{i=1}^{n} x.(y-y')  \times tk$

**Iterasi/Epochs 1**

**[1] Menghitung perubahan $\Delta{b_0}$ dan $\Delta{b_1}$**

Menghitung perubahan $\Delta{b_0}$:

$\Delta{b_0}=\frac{-2}{n} \sum\limits_{i=1}^{n} (y-y') $

$=\frac{-2}{6} [(1-1)+(5-3)+(9-5)+(11-6)+(13-7)+(15-8)]$

$=\frac{-2}{6} 24$

$=-8$


$\Delta {b_1}=\frac{-2}{n} \sum\limits_{i=1}^{n} x.(y-y') $

$=\frac{-2}{6}[1.(1-1)+3.(5-3)5.+(9-5)+6.(11-6)+7.(13-7)+8.(15-8)]$

$=\frac{-2}{6} 154$

$=-51.333$






**[2] Menghitung parameter $b_0$ dan $b_1$ yang baru**

$b_0(baru)=b_0(lama) +\Delta{b_0} \times tk $

$=0-(-8) \times 0.01$

$=0.08$


$b_1(baru)=b_1(lama) +\Delta{b_1} \times tk $

$=1-(-51.333) \times 0.01$

$=1.513$





**[3] Menghitung Nilai Cost/error**

$Cost=\frac{1}{n} \sum\limits_{i=1}^{n} (y_i-y_i')^2$

$=\frac{1}{6} [(1-1)^2 +(5-3)^2 +(9-5^2 )+(11-6)^2 +(13-7)^2 +(15-8)^2]$

$=21.666$

**Iterasi/Epochs 2**

$b_0+b_1 \times X_i$ =  $y'$

$0.08 +1.513(3) = 4.62$
$0.08 +1.513(5) = 7.64666667$

$0.08 +1.513(7) = 10.6733333$

**[1] Menghitung perubahan $\Delta{b_0}$ dan $\Delta{b_1}$**

Menghitung perubahan $\Delta{b_0}$:

$\Delta{b_0}=\frac{-2}{n} \sum\limits_{i=1}^{n} (y-y') $

**[2] Menghitung parameter $b_0$ dan $b_1$ yang baru**

#**Regresi Linier Python From Scratch**

**Membangun Model Regresi Linier**

Pertama, import library python yang dibutuhkan yaitu:
- Numpy
- Pandas
- Matplotlib





````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
%matplotlib inline
````

Untuk memahami konsep, kita akan menggunakan data X dan Y seperti yang kita tunjukkan pada proses penghitungan manual.

````python
x_train=np.array([[1],[3],[5],[6],[7],[8]])
y_train=np.array([1,5,9,11,13,15])
````

Kode dibawah ini dibangun menggunakan kriteria bahwa program akan berhenti ketika jumlah iterasi=n telah dicapai.

````python
def gradient_descent_1(X, y, iterasi, learning_rate = 0.01):

    jumlah_variabel = X.shape[1]                   # array numpy dengan 1 baris dan sejumlah kolom variabel
                                                   # Contoh: X_array([[1,2], [3,4], [5,5], [6,6], [7,2], [8,1]])
    
    bpoly = np.ones(shape=(jumlah_variabel))       # b1, b2, ... b(jumlah variabel)
    b0 = 0                                         # b0 atau bias
    n = X.shape[0]                                 # jumlah baris data X
    
    cost_list = []                                 # list untuk menampung nilai cost setiap iterasi
    iterasi_list = []                              # list untuk menampung nilai n iterasi(1, 2, 3, ..., n)
    b_list=[]                                      # list untuk menampung nilai parameter regresi b0, b1, .. b(jumlah_variabel) 
                                                   # setiap iterasi
    
    # Program akan berjalan hingga sejumlah iterasi=n dicapai
    for i in range(iterasi):   
        y_pred = np.dot(bpoly, X.T) + b0           # Menghitung y_pred pada setiap iterasi
        bpoly_grad = -(2/n)*(X.T.dot(y-y_pred))    # Menghitung perubahan/gradien bpoly setiap iterasi
        b0_grad = -(2/n)*np.sum(y-y_pred)          # Menghitung perubahan/gradien b0 setiap iterasi
        bpoly = bpoly - learning_rate * bpoly_grad # Memperbaharui bpoly setiap iterasi
        b0 = b0 - learning_rate * b0_grad          # Memperbaharui bpoly setiap iterasi
        
        # Menyimpan parameter b dalam list tunggal
        bp=[]
        for jv in range(jumlah_variabel):
            bp1=bpoly[jv]
            bp.append(bp1)
        b=[b0]+bp
        
        cost = np.mean(np.square(y-y_pred))        # MSE (Mean Squared Error)
        
        cost_list.append(cost)                     # Menyimpan nilai cost setiap iterasi
        
        iterasi_list.append(i)                     # Menyimpan nilai iterasi setiap iterasi
        b_list.append(b)                           # Menyimpan nilai parameter b setiap iterasi
        
    return bpoly, b0, cost, b_list, cost_list, iterasi_list  # Ouput
````

````python
bpoly, b0, cost, b_list, cost_list, iterasi_list=gradient_descent_1(x_train, y_train, iterasi=100, learning_rate = 0.01)
````

````python
cost_list
````

````python
plt.xlabel("Iterasi")
plt.ylabel("cost")
plt.title("Cost Vs Iterasi")
plt.plot(iterasi_list,cost_list, color='red', label='Eror')
plt.show()
````

Berikut dalah mendefinisikan fungsi gradient descent yang menghasilkan output vektor $b_0$ dan $b_1$ dan nilai error untuk setiap iterasi. Selanjutnya kita akan melakukan visualisasi berupa animasi yang menunjukkan perbaikan persamaan garis regresi linier untuk setiap iterasi.

````python
def gradient_descent_1(X, y, iterasi, learning_rate = 0.01):

    jumlah_variabel = X.shape[0]                   # array numpy dengan 1 baris dan sejumlah kolom variabel
                                                   # Contoh: X_array([[1,2], [3,4], [5,5], [6,6], [7,2], [8,1]])
    
    bpoly = np.ones(shape=(jumlah_variabel))       # b1, b2, ... b(jumlah variabel)
    b0=0                                         # b0 atau bias
    n = X.shape[0]                                 # jumlah baris data X
    
    cost_list = []                                 # list untuk menampung nilai cost setiap iterasi
    iterasi_list = []                              # list untuk menampung nilai n iterasi(1, 2, 3, ..., n)
    b_list=[]                                      # list untuk menampung nilai parameter regresi b0, b1, .. b(jumlah_variabel) 
                                                   # setiap iterasi
    
    # Program akan berjalan hingga sejumlah iterasi=n dicapai
    for i in range(iterasi):   
        y_pred = np.dot(bpoly, X.T) + b0           # Menghitung y_pred pada setiap iterasi
        bpoly_grad = -(2/n)*(X.T.dot(y-y_pred))    # Menghitung perubahan/gradien bpoly setiap iterasi
        b0_grad = -(2/n)*np.sum(y-y_pred)          # Menghitung perubahan/gradien b0 setiap iterasi
        bpoly = bpoly - learning_rate * bpoly_grad # Memperbaharui bpoly setiap iterasi
        b0 = b0 - learning_rate * b0_grad          # Memperbaharui bpoly setiap iterasi
        
        # Menyimpan parameter b dalam list tunggal
        bp=[]
        for jv in range(jumlah_variabel):
            bp1=bpoly[jv]
            bp.append(bp1)
        b=[b0]+bp
        
        cost = np.mean(np.square(y-y_pred))        # MSE (Mean Squared Error)
        
        cost_list.append(cost)                     # Menyimpan nilai cost setiap iterasi
        
        iterasi_list.append(i)                     # Menyimpan nilai iterasi setiap iterasi
        b_list.append(b)                           # Menyimpan nilai parameter b setiap iterasi
        
    return bpoly, b0, cost, b_list, cost_list, iterasi_list  # Ouput
````

````python
X_train = np.squeeze(np.asarray(X_train))

y_train = np.squeeze(np.asarray(y_train))
````

````python
bpoly, b0, cost, b_list, cost_list, iterasi_list = gradient_descent_1(X_train, y_train, iterasi=100, learning_rate = 0.01)
````

**Evaluasi Model**

KIta perlu mengukur tingkat ketepatan(akurasi)model yang telah diperoleh. Terdapat banyak metode untuk melakukan hal ini., namun pada artikel ini kita menggunakan **Root Mean Square Error** dan **Koefisien Determinasi($R^2$ Skor)**

$RMSE = \sqrt{\frac{1}{n} \sum\limits_{i=1}^{n} (y_i- \hat{y_i})^2}$

$\hat{y_i}$ adalah nilai hasil prediksi ke-i. Berikut adalah penulisan kode untuk RMSE. 

````python
rmse = 0
for i in range(len(X_test)):
  y_pred = b0 +bpoly* X_test[i]
  rmse += (y_test[i]- y_pred) **2

rmse = np.sqrt(rmse/len(X_test))
print(rmse)
````

selanjutnya, mari kita hitung nilai $R^2$ skor untuk mengukur tingkat akurasi dair model linier yang diperoleh, secara metamatis $R^2$ dituliskan sebagai berikut:

$ R^2 = 1- \frac{SSE}{SST} = 1- \frac {\sum\limits(\hat{y_i}- \bar{y})^2} {\sum\limits(y_i- \bar{y})^2}$

SSE adalah jumlah kuadrat galat sedangkan SST jumlah kuadrat total.

$R^2$ skor biasanya memiliki nilai antara 0 hinga 1. Nilai ini juga bisa bernilai negatif jika model regresi benar-benar salah. Di bawah ini adlah kode pemrograman untuk menghitung $R^2$ skor.


````python
SST = 0
SSE = 0
y_pred= b0 +bpoly * X_test
y_rata2= np.mean (y_test)
for i in range (len(X_test)) :
  SST += (y_test[i]-y_rata2) **2
  SSE += (y_test[i] - y_pred[i]) **2

score = 1- (SSE/SST)
print("Koefisien Determinasi :", score)
````
