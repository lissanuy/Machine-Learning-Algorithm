#**Algoritma K-Nearest Neighbor(KNN)**

Algoritma K-Nearest Neighbors adalah algoritma Machine Learning yang melakukan klasifikasi sebuat data didasarkan pada sifat-sifat data diddekatnya. Sebuah uangkapan yang bisa membantu dengan mudah memahami ide dasar dari algoritma ini adalah: "Jika kamu ingin mengetahui sifat sesorang, lihatlah sifat-sifat orang terdekatnya. Sifatnya tidak akan jauh berbeda dari mereka".

Gambar dibawah ini dapat mengilustrasikan hal di atas dengan jelas.

Screenshot (1427).png

Data baru yang dilambangkan dengan simbol segitiga biru dapat diklasifikasikan sebagai kelompok kotak hijau atau bintang merah jambu. Jika dilihat berdasarkan kriteria jumlha k=1 maka data segitiga biru diklasifikasikan ke dalam kelompok kotak hijau. Sedangkan jika dilihat berdasarkan kriteria jumlah k=3 maka data tersebut diklasifikasikan sebagai kelompok bintang merah jambu. Hal yang sama juga dapat diterapkan untuk k=9, Sebenarnya, kita telah melakukan klasifikasi berdasarkan jarak. Kedekatan yang dimaksud ketika k=1, k=3, dan k=9, kita selalu mengukur kedekatan antara data baru(segitiga biru) dengan data-data berlabel kelompok sebelumya. Berdasarkan ide dasr diatas, maka diturunkanlah dalam sebuah algoritma untuk menentukan kelompok data dari sebuah inputan data baru(segitiga biru).

#**Algoritma KKN**

1. Menentukan paramter k(jumlah tetangga paling dekat)
2. Menghitung jarak eucliden objek terhadap data training yang diberikan.
3. Mengurutkan hasil no 2 secara asdencing (berurutan dan nilai tinggi ke rendah)
4. Mengumpulkan kategori data(klasifikasi nearest neighbor berdasarkan nilai k)
5. Dengan menggunakan kategori nearest neighbor yang paling mayoritas maka adapat diprediksi kategori objek.

#**Contoh Perhitungan K-Nearest Neighbort**

Diketahui data $x_1$ dan $x_2$ adalah sebagai berikut:


$x_1$  |$x_2$   |Kelas
-------|------- |-----

1  | 2 | 0
3  | 2 | 0

Menentukan parameter k(jumlah tetangga paling dekat)

k=3

**Langlah 2:**

Menghitung jarak eucliden objek terhadap data training yang diberikan.

Formula jarak Euclidean:
$d_ij=\sqrt{(x1_i-x1_j)^2+ (x2_i-x2_j)^2} untuk i ,j=1,2,3,..n$

$x_1$  |$x_2$   |Jarak Euclidean dengan (9.10)
-------|------- |-----

11 | 13|$\sqrt{(9-11)^2 +(10-13)^2=3.60}$
14 | 15|$\sqrt{(9-14)^2 +(10-15)^2=7.07}$






**Langkah 3:**
Mengurutkan hasil langah 2 secara ascending(berurutan dari nilai tinggi ke rendah)

$x_1$  |$x_2$   |Jarak Euclidean dengan (9.10) |Urutan(Jarak Terpendek)

11 | 13|$\sqrt{(9-11)^2 +(10-13)^2=3.60}$ | 2
14 | 15|$\sqrt{(9-14)^2 +(10-15)^2=7.07}$| 3


**Langkah 4:**

Mengumpulkan kategori data (Klasifikasi nearest neighbor berdasarkan nilai k)

Kelas data 3 terdekat(k=3)

(12,11)=1,(11,13)=1,(14,15)=1

**Langkah 5:"**

Dengan menggunakan kategori nearest neighbor yang paling mayoritas maka dapat diprediksikan kategori objek.

kelas data untuk k=3

'0'=0, '1'=3

Maka disimpulkan bahwa data (9,10) diklasifikasikan ke kelas '1'.









#**KNN Python From Scratch**

**Membangun Model KNN**

Pertama, Import library python yang dibutuhkan yaitu:
- Numpy
-Pandas
-Matplotlib




````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
X=np.array([[1,1],[2,2],[3,2],[12,11],[11,13],[14,15]])
y=np.array([0,0,0,1,1,1])
x_baru=np.array([9,10])
````

````python
X
````

````python
y
````

````python
x_baru
````

**langkah 1:**

Menentukan parameter k(jumlah tetangga paling dekat)

````python
k=3
````

**Langkah 2:**
Menghitung kuadrat jarak eucliden objek terhadap data training yang diberikan.

````python
def jarak_euclidean(x,y):
  return np.sqrt(sum(pow(a-b, 2)for a, b in zip(x,y)))
````

````python
jarak=[]
for x_i in X:
  jarak_euclid=jarak_euclidean(x_i,x_baru)
  jarak.append(jarak_euclid)
  print(jarak)
print(jarak)
````

**Langkah 3:** Mengurutkan hasil langkash 2 secara ascending(berurutan dari nilai tinggi ke rendah)


````python
urut=np.sort(jarak)
list_urut=[]
for i in range (len(urut)):
  list_urut.append(urut[i])
print(list_urut)
````

````python
urutan=[]
for i in list_urut:
  urtn=jarak.index(i)
  urutan.append(urtn)
print(urutan)
````

atau dengan sederhana dapat mengunakan ```np.argsort``` seperti berikut:

````python
np.argsort(jarak)
````

**Langkah 4:**

Mengumpulkan kategori data(Klasifikasi nearest neighbor berdasarkan nilai k)

Kelas data 3 data terdekat(k=3)


````python
ambil=urutan[:k]
print(y[ambil])
````

**Langkah 5:**

Dengan mengggunakan kategori nearest neighbor yang paling mayoritas maka dapat diprediksikan kategori objek.

````python
from collections import Counter
kelas_paling_banyak = Counter(y[ambil]).most_common(1)
kelas_paling_banyak
````

````python
kelas_paling_banyak[0][0]
````

````python
# Plot data
df=pd.DataFrame(X,columns =['x1','x2'])
colormap=np.array(['g','b'])
label=['A','A','A','B','B','B']

# Plot Titik Data
plt.scatter(df['x1'],df['x2'],color=colormap[y])
plt.scatter(x_baru[0],x_baru[1],color='r',label='Data Baru')

# Label sumbu x
plt.xlabel('x_1')

# Label sumbu y
plt.ylabel('x_2')

# Menampilkan Legenda
plt.legend()

# Menampilkan Gambar Hasil Plot
plt.show()
````

Maka dapat disimpulkan bahwa data(9,10) diklasifikasikan ke kelas '1'

#**Implementasi Model KKN Pada Data Real**

Kita akan menggunakan dataset Iris yang berada di sklearn dataset

````python
from matplotlib.colors import ListedColormap
from sklearn import datasets
from sklearn.model_selection import train_test_split
iris = datasets.load_iris()
iris
````

Memilih dataset X(iris.data) dan dataset y(iris.target)


````python
X, y = iris.data, iris.target
````

Membagi dataset menjadi data latih dan data uji

````python
X_latih, X_uji, y_latih, y_uji = train_test_split(X, y, test_size=0.20, random_state=1234)
````

Menuliskan algoritma KNN dalam bentuk fungsi

````python
def _knn(x):
 # Menghitung jarak antara x baru dengan semua sampel di data latih
 jarak = [jarak_euclidean(x, x_latih) for x_latih in X_latih]
 # Mengurutkan data dan mengambil sejumlah k objek
 k_idx = np.argsort(jarak)[: k]
 # Mengambil kelas dari k tetangga terdekat dari data latih
 k_neighbor_labels = [y_latih[i] for i in k_idx]
 # Mengambil kelas yang paling sering muncul
 kelas_paling_banyak = Counter(k_neighbor_labels).most_common(1)
 return kelas_paling_banyak[0][0]
 
def KNN(X_latih, y_latih, k, x_baru):
 y_pred= [_knn(x_i) for x_i in x_baru]
 return np.array(y_pred)
````

````python
y_pred=KNN(X_latih, y_latih, k, X_uji)
y_pred
````

#**Evaluasi Model**

Evaluasi model digunakan untuk melihat seberapa baik metode ini bekerja pada data uji yang diberikan. Salah satu metrik yang dapat digunakan adalah akurasi. Metrik ini menghitung jumlah hasil prediksi yang sesuai dengan label sesungguhnya kemudian dibagi dengan jumlah keseluruhan label. Menggunakan bahasa python akurasi dapat dituliskan sebagai berikut.



````python
def akurasi(y_asli, y_pred):
 akurasi = np.sum(y_asli == y_pred) / len(y_asli)
 return akurasi
````

````python
akurasi(y_uji , y_pred)
````

Diperoleh hasil akurasi 1.0 artinya metode KNN dapat melakukan prediksi dengan tepat untuk semua data uji yang digunakan
